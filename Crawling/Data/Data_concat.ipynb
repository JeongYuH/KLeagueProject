```python
import pandas as pd
import os
```

```python
list = os.listdir()
data = pd.DataFrame()
```

```python
# 데이터 결합하기
for li in list[:-1]:
    a = pd.read_csv(f'{li}', index_col='Unnamed: 0')
    print(a)
    data = pd.concat([data, a])
```

```python
# 데이터 년도, 디비전 기준으로 정렬
data = data[:].sort_values(['년도', '리그'])
```

```python
# 결합된 데이터 저장
data.to_csv('2020_2023.csv')
```
